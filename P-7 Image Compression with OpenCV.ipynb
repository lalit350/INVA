```python
import cv2
from google.colab.patches import cv2_imshow
from PIL import Image
import matplotlib.pyplot as plt
# Load the image
image_path = ('/content/nature1.jpg')
image_path2 = ('/content/nature2.jpg')
image = cv2.imread(image_path)
cv2_imshow(image)
image2 = cv2.imread(image_path2)
cv2_imshow(image2)






# Save the image with lossless compression (PNG)
cv2.imwrite('image_lossless.png', image)
cv2.imwrite('image_lossless2.png', image2)






# Save the image with lossy compression (JPEG)
cv2.imwrite('image_lossy.jpg', image, [cv2.IMWRITE_JPEG_QUALITY, 10])
cv2.imwrite('image_lossy2.jpg', image2, [cv2.IMWRITE_JPEG_QUALITY, 10])






# Load the compressed images
image_lossless = cv2.imread('image_lossless.png')
image_lossy = cv2.imread('image_lossy.jpg')

# Plot the images
plt.figure(figsize=(10, 5))
# Original Image
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')
# Lossless Compressed Image
plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(image_lossless, cv2.COLOR_BGR2RGB))
plt.title('Lossless Compression (PNG)')
plt.axis('off')
# Lossy Compressed Image
plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(image_lossy, cv2.COLOR_BGR2RGB))
plt.title('Lossy Compression (JPEG)')
plt.axis('off')
plt.show()

image_lossless2 = cv2.imread('image_lossless2.png')
image_lossy2 = cv2.imread('image_lossy2.jpg')
# Plot the images
plt.figure(figsize=(10, 5))
# Original Image
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(image2, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')
# Lossless Compressed Image
plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(image_lossless2, cv2.COLOR_BGR2RGB))
plt.title('Lossless Compression (PNG)')
plt.axis('off')
# Lossy Compressed Image
plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(image_lossy2, cv2.COLOR_BGR2RGB))
plt.title('Lossy Compression (JPEG)')
plt.axis('off')
plt.show()








import os

# Get file sizes
original_size = os.path.getsize(image_path)
lossless_size = os.path.getsize('image_lossless.png')
lossy_size = os.path.getsize('image_lossy.jpg')

print("Shape of the Original Image:", image.shape)
print("Shape of the Lossless Compressed Image:", image_lossless.shape)
print("Shape of the Lossy Compressed Image:", image_lossy.shape)

print(f"\nOriginal Image Size: {original_size / 1024:.2f} KB")
print(f"Lossless Compressed Image Size (PNG): {lossless_size / 1024:.2f} KB")
print(f"Lossy Compressed Image Size (JPEG): {lossy_size / 1024:.2f} KB")

original_size2 = os.path.getsize(image_path2)
lossless_size2 = os.path.getsize('image_lossless2.png')
lossy_size2 = os.path.getsize('image_lossy2.jpg')

print("\nShape of the Original Image:", image2.shape)
print("Shape of the Lossless Compressed Image:", image_lossless2.shape)
print("Shape of the Lossy Compressed Image:", image_lossy2.shape)

print(f"\nOriginal Image Size: {original_size2 / 1024:.2f} KB")
print(f"Lossless Compressed Image Size (PNG): {lossless_size2 / 1024:.2f} KB")
print(f"Lossy Compressed Image Size (JPEG): {lossy_size2 / 1024:.2f} KB")
```
